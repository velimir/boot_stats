```python
%matplotlib inline

import pandas as pd
import seaborn as sns

sns.set()
```

```python
df = pd.read_csv('stats.csv')
df['first_call'] = df['init'].isna()
```

```python
ax = sns.boxplot(x=df[df['mode'] == 'interactive']['init'])
```

```python
ax = sns.boxplot(x=df[df['mode'] == 'embedded']['init'])
```

```python
ax = sns.boxplot(x="mode", y="init", data=df)
```

```python
ax = sns.boxplot(x="mode", y="duration", hue="first_call", data=df)
```
